```python
print("\n" + "="*50)
print("🔍 SENTIMENT CLASSIFICATION ANALYSIS")
print("="*50)

def classify_reviews_batch(reviews_data, batch_size=5):
    """Classify reviews in batches for better performance"""
    results = []

    for i in range(0, len(reviews_data), batch_size):
        batch = reviews_data[i:i+batch_size]

        # Prepare batch text
        batch_text = "\n".join([
            f"Review {idx+1} (Product: {review['product_category']}): {review['review_text']}"
            for idx, review in enumerate(batch)
        ])

        # Classification prompt
        classification_prompt = f"""
        Analyze and classify these customer reviews with the following format:

        For each review, provide:
        - Sentiment: [Positive/Negative/Mixed]
        - Product Issues: [List main issues mentioned, or 'None' if no issues]
        - Product Strengths: [List main strengths mentioned, or 'None' if no strengths]
        - Overall Category: [Performance/Design/Value/Reliability/User Experience]

        Reviews to analyze:
        {batch_text}

        Please provide structured analysis for each review.
        """

        # Model parameters for classification
        parameters = {
            "top_k": 5,
            "top_p": 0.9,
            "max_tokens": 512,
            "min_tokens": 50,
            "repetition_penalty": 1.1
        }

        try:
            response = granite_llm.invoke(classification_prompt, parameters=parameters)
            results.append({
                'batch_number': i//batch_size + 1,
                'reviews': batch,
                'classification_result': response
            })
            print(f"✅ Processed batch {i//batch_size + 1}")
        except Exception as e:
            print(f"❌ Error processing batch {i//batch_size + 1}: {e}")

    return results
```

```python
classification_results = classify_reviews_batch(customer_reviews, batch_size=4)
```

```python
for result in classification_results:
    print(f"\n📋 BATCH {result['batch_number']} CLASSIFICATION RESULTS:")
    print("-" * 60)
    print(result['classification_result'])
    print("-" * 60)
```

```python
print("\n" + "="*50)
print("📝 BUSINESS INSIGHTS SUMMARIZATION")
print("="*50)

def generate_business_summary(reviews_data):
    """Generate comprehensive business summary from all reviews"""

    # Prepare all reviews text
    all_reviews_text = "\n".join([
        f"Product {review['product_category']} (Rating: {review['rating']}/5): {review['review_text']}"
        for review in reviews_data
    ])

    # Business summary prompt
    summary_prompt = f"""
    As a business analyst, provide a comprehensive summary of customer feedback data for an e-commerce platform.

    Structure your analysis as follows:

    ## EXECUTIVE SUMMARY
    [2-3 sentences overview of overall customer sentiment and key findings]

    ## KEY INSIGHTS BY PRODUCT CATEGORY
    [Analyze patterns by product type - smartphones, laptops, headphones, etc.]

    ## CRITICAL ISSUES IDENTIFIED
    [List top 3-5 most common customer complaints across all products]

    ## POSITIVE FEEDBACK HIGHLIGHTS
    [List top 3-5 most appreciated features/aspects]

    ## BUSINESS RECOMMENDATIONS
    [Provide 4-5 actionable recommendations based on the analysis]

    ## PRIORITY ACTION ITEMS
    [List immediate actions needed based on customer feedback]

    Customer Reviews Data:
    {all_reviews_text}
    """

    # Model parameters for summarization
    parameters = {
        "top_k": 10,
        "top_p": 0.95,
        "max_tokens": 800,
        "min_tokens": 300,
        "repetition_penalty": 1.2
    }

    try:
        response = granite_llm.invoke(summary_prompt, parameters=parameters)
        return response
    except Exception as e:
        print(f"❌ Error generating business summary: {e}")
        return None

# Generate business insights
business_summary = generate_business_summary(customer_reviews)

if business_summary:
    print("🎯 BUSINESS INTELLIGENCE SUMMARY:")
    print("=" * 70)
    print(business_summary)
    print("=" * 70)
```

```python
print("\n" + "="*50)
print("🔬 ADVANCED ANALYTICS")
print("="*50)

# Extract key metrics
def calculate_key_metrics(df):
    metrics = {
        'total_reviews': len(df),
        'average_rating': df['rating'].mean(),
        'positive_reviews': len(df[df['rating'] >= 4]),
        'negative_reviews': len(df[df['rating'] <= 2]),
        'most_reviewed_category': df['product_category'].mode()[0],
        'latest_review_date': df['date'].max(),
        'rating_distribution': df['rating'].value_counts().to_dict()
    }
    return metrics

metrics = calculate_key_metrics(df_reviews)

print("📊 KEY PERFORMANCE METRICS:")
print("-" * 40)
for key, value in metrics.items():
    if key != 'rating_distribution':
        print(f"{key.replace('_', ' ').title()}: {value}")

print(f"\nRating Distribution: {metrics['rating_distribution']}")

# Word frequency analysis
def analyze_text_patterns(reviews_text):
    """Analyze common words and patterns in reviews"""
    all_text = ' '.join(reviews_text).lower()

    # Remove common stop words and clean text
    stop_words = {'the', 'and', 'is', 'it', 'to', 'of', 'a', 'in', 'for', 'with', 'but', 'very', 'this', 'that'}
    words = re.findall(r'\b\w+\b', all_text)
    filtered_words = [word for word in words if word not in stop_words and len(word) > 2]

    return Counter(filtered_words).most_common(15)

common_words = analyze_text_patterns(df_reviews['review_text'])
print(f"\n🔤 MOST COMMON WORDS IN REVIEWS:")
print("-" * 40)
for word, count in common_words:
    print(f"{word}: {count}")
```
